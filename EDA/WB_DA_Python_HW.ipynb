**Четвертая гипотеза:** В некоторых штатах (северных) чаще покупают одежду осенью (перед зимой утепляются)

**Вывод:** хотя для некоторых холодных штатов частота покупок осенью действительно выше, нельзя сказать что гипотеза на 100% верна из-за малого кол-ва данных

<h1>Задание 3</h1> 

У нас есть полный каталог всех случаев, когда кто-то проклинал или истекал кровью в фильме Квентина Тарантино. https://github.com/fivethirtyeight/data/tree/master/tarantino

Определите:

* число проклятий, летальных исходов и их соотношение по фильмам 

* частоту употребления конкретных проклятий 

* распределение времени между проклятиями 

```python
df = pd.read_csv('tarantino.csv')
df
```

```python
# Общее число проклятий и смертей:
df.type.value_counts()
```

```python
translator = {'word' : 'проклятий',
             'death' : 'смертей'}
for type in ['word', 'death']:
    pivot_table = df[df.type == type].pivot_table(
    index = 'movie',
    columns = 'type',
    values = 'minutes_in',
    aggfunc = 'count'
    )
    fig = plt.figure(figsize = (4, 6))
    sns.heatmap(pivot_table, annot = True, cmap = 'crest', fmt = '.0f')
    plt.xlabel('')
    plt.title(f'Число {translator[type]} по фильмам')
    plt.show();
```

Посмотрим на частоту употребления конкретных проклятий:

```python
(df.word.value_counts(normalize = True)*100).reset_index()
```

```python
# Вот самые частые:
fig = plt.figure(figsize = (12 , 6))
words = ((df.word.value_counts(normalize = True)*100).reset_index()).head(10)
sns.barplot(x = words.word, y = words.proportion)
```

Посмотрим на распределение времени между проклятиями:

```python
data = df[df.type == 'word']
```

```python
data['time_diff'] = data.groupby('movie')['minutes_in'].diff()
data = data[~data.time_diff.isna()]
```

Посмотрим на распределение времени между проклятиями

```python
sns.histplot(data = data, x = 'time_diff', bins = 30);
```

Обрежем выбросы справа и слева:

```python
sns.histplot(data = data[(data.time_diff <= data['time_diff'].quantile(0.95)) &
              (data.time_diff >= data['time_diff'].quantile(0.05))], x = 'time_diff', bins = 30, kde = True);
```

Также можем посмотреть на распределение, относительно разных фильмов:Э

```python
fig = plt.figure(figsize = (15, 12))
sns.histplot(data = data[(data.time_diff <= data['time_diff'].quantile(0.95)) &
              (data.time_diff >= data['time_diff'].quantile(0.05))], x = 'time_diff', bins = 30, kde = True, hue = 'movie');
```
